```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../../Data/tiger.jpg')

# Get image dimensions
h, w = img.shape[0:2]

# Add padding to the image
base_size = h + 20, w + 20, 3
base = np.zeros(base_size, dtype=np.uint8)

# Draws a black outline inside the base 30 pixels deep
# It is an outline due to (0, 0) as start and (w+20, w+20) as end
# (0, 0, 0) give the RGB color
# Is not needed in this example as np.zeros give black padding
cv2.rectangle(base, (0, 0), (w + 20, h + 20), (0, 0, 0), 130)

# Puts image into the base rectangle
base[10:h+10, 10:w+10] = img
```

```python
# This allows us to display the images correctly
base = cv2.cvtColor(base, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
ax1.imshow(img)
ax1.axis('off')
ax1.set(title='Original')

ax2.imshow(base)
ax2.axis('off')
ax2.set(title='Padded')
```
