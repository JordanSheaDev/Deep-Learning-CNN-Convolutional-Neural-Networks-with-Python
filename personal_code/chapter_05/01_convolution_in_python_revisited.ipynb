```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import cv2
```

```python
img = plt.imread('../../Data/bird.png')
img = img[:,:,:3]
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
mask_blur = np.ones((19,19))/(19*2)
```

```python
img_blur = signal.convolve2d(img_gray,
                             mask_blur,
                             boundary='symm',
                             mode='same')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,15))
ax1.imshow(img_gray, cmap='gray')
ax1.axis('off')
ax1.set(title='Original')
ax2.imshow(img_blur, cmap='gray')
ax2.axis('off')
ax2.set(title='Blur')
```
